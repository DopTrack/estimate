# Gravity Sensitivity exercise 
This Jupyter Notebook can be used to calculate different components of the grvaity field model XGM2019_e spherical harmonics model. We will take a look at the different sensitivities of the components, at different heights, and bandpass truncations to assess what the best gravity components is for your study. To handle the data in this tutorial, we use an open source python package SHtools (https://shtools.github.io/SHTOOLS/index.html). 

### UNITS and conventions
To better inspect the anomalies in the gravity field that depict local mass variation in the subsurface, usually the normal gravity field or the gravity due to the ellipsoid is substracted
We use the WGS84 ellipsoid to calculate the values for the unnormalized SH coefficients:
    - C00  = 1
    - C20  = -1.08262982131e-3
    - C40  = 2.37091120053e-6
    - C60  = -6.08346498882e-9
    - C80  = 1.42681087920e-11

Gravity potential unit is m^2/s^2 and when plotted on the Reference ellipsoid can beconverted to Geoid (m) by dividing by the normal gravity (GM/R^2) or 9.81 m/s^2. (Brun's formula)

Gravity data has to be imported as geographical coordinates (x,y,value) with mGal as unit. Depth information has to be defined the same, depth informationin m, z negative downwards. The algorithm will perform automatically an UTM conversion. 

Gravity gradient component are the second derivative of the gravity potential. The gravity gradient is a tensor and has 6 independent component.

# Data acquisition

First we need to extract the data and plot the resulting fields that we are going to use.

```python
# -*- coding: utf-8 -*-
"""
Created on Tue Feb 16 2021

@author: Bart Root
"""

#import matplotlib
#matplotlib.use('TkAgg')
import matplotlib.pyplot as plt
import pyshtools as pysh
from pyshtools import constants
import pyshtools.datasets.Earth.Earth2014 as TOPO
from cartopy import crs as ccrs
import numpy as np 

pysh.utils.figstyle(rel_width=1)
#% config InlineBAckend.figure_format = 'retina'

#clm = pysh.datasets.Mars.GMM3()
clm = pysh.SHGravCoeffs.from_file('XGM2019e_2159.gfc', format='icgem')
#clm = pysh.SHGravCoeffs.from_file('XGM2019.gfc', format='icgem') # lmax not more than 760

clm.info()
```

```python
# Reference the gravity field model to the WGS84 ellipsoid.
a = constants.Earth.wgs84.a.value
f = constants.Earth.wgs84.f.value
u0 = constants.Earth.wgs84.u0.value
OO = constants.Earth.wgs84.omega.value
clm.set_omega(constants.Earth.wgs84.omega.value)

# outputing the normal gravity coefficients.
print( 'a = ', a, ' m')
print( 'f = ', f)
print( '1/f = ', 1/f)
print( 'u0 = ', u0, ' m^2/s^2')
print( 'OO = ', OO, ' rad/s')

# plotting the the power spectrum of the data in 1D and 2D
fig1, ax1 = clm.plot_spectrum(function='geoid', show=False)
fig2, ax2 = clm.plot_spectrum2d(function='total', show=False, errors=False)

clm.info()
```

```python
# Determine the resolution of the data you are going to work with
LMAX = 200

# gravity disturbance
grav = clm.expand(lmax=LMAX, a=a, f=f)
fig3, ax3 = grav.plot_total(cmap='RdBu_r',
                            cmap_limits=[-200, 200],
                            show=False)

ellipsoid = clm.geoid(u0, lmax=LMAX)
geoid = clm.geoid(u0, a=a, f=f, lmax=LMAX)

# geoid and ellipsoid 
fig4, ax4 = geoid.plot(cmap='RdBu_r',
                       cmap_limits=[-100, 100], 
                       show=False)
fig5, ax5 = ellipsoid.plot(cmap='RdBu_r', 
                           show=False)
```

```python
# Import the topography for the topogrpahic reduction
shape = TOPO.ret(lmax=LMAX)  # read only the first 760 degrees from the file.
shape_grid = shape.expand(grid='DH2')
topo_grid = shape_grid / 1.e3

fig6, ax6 = topo_grid.plot(colorbar='bottom',
                           cb_label='Topography, km',
                           cmap='terrain',
                           cmap_limits=[-2, 8],
                           show=False)
#shape.info()
```

```python
# For the Bouguer correction done properly in SHTools, the radius of the sphere needs to be added to the topography
topo_grid2 = shape_grid + a

# expand the grid file back to SH coefficients
shapev2 = topo_grid2.expand(lmax_calc=LMAX)

# make the Bourguer correction from the shape
bc = pysh.SHGravCoeffs.from_shape(shapev2,
                                  rho=2650.,
                                  gm=clm.gm,
                                  lmax=LMAX)
# set reference sphere equal to clm reference sphere
bc = bc.change_ref(r0=clm.r0)

# Construct the topogrpahic reduced gravity field (SH coefficients)
bouguer = clm.pad(lmax=LMAX) - bc
bouguer.info()
```

```python
##### Projection information #########

PROJECTION = ccrs.PlateCarree(central_longitude=-180.)

roi = [50, 74, 38, 50] # local area
#roi = [-180, 180, -90, 90] # global area
```

```python
# Expand Coefficients into a grid
bouguer_grid = bouguer.expand(lmax=LMAX,a=a,f=f)

fig7, ax7 = bouguer_grid.plot_total(projection=PROJECTION,
                            cmap='RdBu_r',
                            cmap_limits=[-800, 800], 
                            cb_triangles='both',
                            show=False)
#ax7.axis(roi)
```

```python
# For the gradient plots the normal gravity needs to be subtracted from the data. 
Vlm = bouguer.copy()
Vlm.set_omega(0) # set the rotation to zero, to remove centrifugal accelaration

# Set coefficients C00, C20, C40, C60, C80 such, that they are reduced by the WGS84 ellisoidal signal
aa = Vlm.to_array()
Vlm.set_coeffs(ls=0,ms=0,values=0.)
Vlm.set_coeffs(ls=2,ms=0,values=aa[0,2,0]+1.08262982131e-3/np.sqrt(5))
Vlm.set_coeffs(ls=4,ms=0,values=aa[0,4,0]-2.37091120053e-6/np.sqrt(9))
Vlm.set_coeffs(ls=6,ms=0,values=aa[0,6,0]+6.08346498882e-9/np.sqrt(13))
Vlm.set_coeffs(ls=8,ms=0,values=aa[0,8,0]-1.42681087920e-11/np.sqrt(17))
```

```python
# calulate the tensor components
bouguer_grid = Vlm.expand(lmax=LMAX, a=a)
tensor = Vlm.tensor(lmax=LMAX, a=a)


# plot the fields
# try out different components with:
# .plot_rad, .plot_theta, .plot_phi, .plot_pot, .plot_total
fig9, ax9 = bouguer_grid.plot_rad(projection=PROJECTION,
                            cmap='RdBu_r',
                            cmap_limits=[-400/1e5, 400/1e5], 
                            cb_triangles='both',
                            show=False)
#ax9.axis(roi)


# plot tensor components
# try out different components with:
# .plot_vxx, .plot_vyy, .plot_vzx, .plot_vzy, .plot_vxy
fig10, ax10 = tensor.plot_vzz(projection=PROJECTION,
                          cmap='RdBu_r', 
                          #cmap_limits=[-25, 25], 
                          show=False)
#ax10.axis(roi)
```

```python
# set truncation limits for coefficients

set_a = clm.pad(lmax=LMAX)-bc
set_a.set_omega(0) # remove centrifugal acceleration.

# minimum and maximum truncation
lmin = 10
lmax = 359
for lsh in range(0, lmin):
    for msh in range(-lsh, lsh+1):
        #print('Degree:      ', lsh)
        #print('Order: ', msh)
        set_a.set_coeffs(ls=[lsh],
                         ms=[msh],
                         values=0.)
    
# synthesize the different components.
set_a_grid = set_a.expand(lmax=lmax,a=a)
set_a_tensor = set_a.tensor(lmax=lmax,a=a)
```

```python
# colormap settings
clims = 400/1e5

# plot the different vector components
fig3_1, ax3_1 = set_a_grid.plot_rad(projection=PROJECTION,
                            cmap='RdBu_r',
#                            cmap_limits=[-clims, clims], 
                            cmap_limits=[-0, 500/1e5], 
                            cb_triangles='both',
                            show=False)
ax3_1.axis(roi)

fig3_2, ax3_2 = set_a_grid.plot_theta(projection=PROJECTION,
                            cmap='RdBu_r',
                            cmap_limits=[-clims, clims], 
                            cb_triangles='both',
                            show=False)
ax3_2.axis(roi)

fig3_3, ax3_3 = set_a_grid.plot_phi(projection=PROJECTION,
                            cmap='RdBu_r',
                            cmap_limits=[-clims, clims], 
                            cb_triangles='both',
                            show=False)
ax3_3.axis(roi)
```

```python
# sensitivity of the measured gravity to certain mass sources in the subsurface.
height = 250e3    # height of the observations done

depth = 20e3      # Depth of the mass source (crust typically 30 l, Lith-mantle 100-300 km)

# radius Earth
R = 6378000
r = R+height
rd = R-depth

# angular distance between the observation and mass source
phi = np.arange(0, 15, 0.1)/180*np.pi # angular distance between comp and integration point

# caomputing parameters from Martinec (2014)
x = np.cos(phi)
t = rd/r
g = np.sqrt(1+t**2 - 2*t*x)

# Gravity vector mass-density Green's functions
Kr = -(1-t*x)/(g**3)
KO = np.sqrt(1-x**2)*(t)/(g**3)

# plot figure
plt.figure()
ax_4_1 = plt.plot(phi/np.pi*180, -Kr/np.max(-Kr), label = "Kr") 
ax_4_1 = plt.plot(phi/np.pi*180, KO/np.max(KO), label = "KO")
#plt.axis([0,15,-10,70])
plt.xlabel('Phi (degree)') 
plt.ylabel('Isotropic kernels') 
plt.title('Vector Kernel functions') 
plt.legend() 
```

```python
#Gravity gradient tensor mass-density Green's functions

Krr = -1/g**3 + (3*(1-t*x)**2)/(g**5)
KOO = 0.5*(1-x**2)*(3*t**2)/(g**5)
KrO = -np.sqrt(1-x**2)*(3*t*(1-t*x))/(g**5)

# plot figure
plt.figure()
ax_4_2 = plt.plot(phi/np.pi*180, Krr/np.max(Krr), label = "Krr") 
ax_4_2 = plt.plot(phi/np.pi*180, KOO/np.max(KOO), label = "KOO") 
ax_4_2 = plt.plot(phi/np.pi*180, -KrO/np.max(-KrO), label = "KrO") 
#plt.axis([0,15,-1500,30000])
plt.xlabel('Phi (degree)') 
plt.ylabel('Isotropic kernels') 
plt.title('Gradient Kernel functions') 
plt.legend()
```

```python
depth_sens = np.arange(0, 1000, 0.1)*1e3

dKv2 = (R/(R-depth_sens))**(2+2)
dKv2 = dKv2/dKv2[-1]
dKv7 = (R/(R-depth_sens))**(7+2)
dKv7 = dKv7/dKv7[-1]
dKv30 = (R/(R-depth_sens))**(30+2)
dKv30 = dKv30/dKv30[-1]
dKv90 = (R/(R-depth_sens))**(90+2)
dKv90 = dKv90/dKv90[-1]

# plot figure
plt.figure() 
ax_5_1 = plt.plot( dKv2, np.max(depth_sens/1e3)-depth_sens/1e3, label = "2") 
ax_5_1 = plt.plot( dKv7, np.max(depth_sens/1e3)-depth_sens/1e3, label = "7") 
ax_5_1 = plt.plot( dKv30, np.max(depth_sens/1e3)-depth_sens/1e3, label = "30") 
ax_5_1 = plt.plot( dKv90, np.max(depth_sens/1e3)-depth_sens/1e3, label = "90") 
#plt.axis([0,1,0,100])
plt.xlabel('Amplitude') 
plt.ylabel('Depth (km)') 
plt.title('Gradient Kernel functions') 
plt.legend()
plt.gca().invert_yaxis()
```
